```python
import glob
import numpy as np
import matplotlib as mlib
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.lines as mlines
import re
import os
from scipy import interpolate
from numpy import genfromtxt
from numpy import loadtxt
```

```python
# Constants

mh=125
```

```python
# Supporting functions
    
# Simple spline interpolation
def interpolateByX(file,degree,smoothness):
    inputdata=genfromtxt('{}'.format(file), comments="#", delimiter='\t')
    
    return interpolate.UnivariateSpline(inputdata[:,0],inputdata[:,1],k=degree,s=smoothness)


# Interpolation
def arrInterpolate(arr, degree, smoothness):
    Interp = interpolate.UnivariateSpline(arr[:,0],arr[:,1],k=degree,s=smoothness)
    x = np.linspace(arr[0,1],arr[-1,0],200)
    arrInterp = Interp(x)
    return np.vstack((x,arrInterp)).T
```

```python
# Getting DOF(T)

DOFInterp=interpolateByX('sterile_spectrum/mydata/DOF.dat',2,0)

xnew = np.logspace(np.log(2*10**(-6)), np.log(300), 1000)

plt.xlabel('$T$ [GeV]')
plt.ylabel('$g_S^*$')


plt.loglog(xnew, DOFInterp(xnew))

plt.xlim( (10**(-6), 1000))
plt.ylim((1,150))

plt.show()
```

```python
# Conversion functions

# Converting zeta to q (from https://arxiv.org/pdf/1609.01289.pdf):
def qfromzeta(arrzeta,r):
    return arrzeta*(DOFInterp(mh/r)/DOFInterp(mh))**(1/3)
```

```python
# spectrum conversion for sterile neutrinos (https://arxiv.org/pdf/1609.01289.pdf) to f(q)

def spectrumConversionSterile(inputfile,r):
    inputdata=genfromtxt('{}'.format(inputfile)+'.dat', comments="#", delimiter='\t')

    plt.loglog(inputdata[:,0],inputdata[:,1])
    plt.xlabel('$\zeta$')
    plt.ylabel('$\zeta^2 f(\zeta)$')
    plt.show()

    converteddata=np.transpose(np.array([qfromzeta(inputdata[:,0],r),inputdata[:,1]/inputdata[:,0]**2]))

    plt.loglog(converteddata[:,0],converteddata[:,1])

    plt.xlabel('$q$')
    plt.ylabel('$f(q)$')

    #plt.xlim( (10**(-6), 1000))
    #plt.ylim((1,150))

    plt.show()

    with open("sterile_{}_converted.dat".format(inputfile), 'w') as f:
                        f.write('#source: https://arxiv.org/pdf/1609.01289.pdf' + '\n' \
                                +'#q=p/T f(q) for r={}'.format(r) + '\n')
                        np.savetxt(f, converteddata, fmt='%4e')
```

```python
# sterile neutrinos WIMP spectrum conversion to f(q)

spectrumConversionSterile("sterile_spectrum/mydata/spectum_mS60_WIMP",426)
```

```python
# sterile neutrinos FIMP spectrum conversion to f(q)

spectrumConversionSterile("sterile_spectrum/mydata/spectum_mS60_FIMP",259)
```

```python
# spectrum conversion for sterile neutrinos (https://arxiv.org/pdf/1609.01289.pdf) to f(q)

def spectrumConversionScalar(inputfile,r):
    inputdata=genfromtxt('{}'.format(inputfile)+'.dat', comments="#", delimiter='\t')

    plt.plot(inputdata[:,0],inputdata[:,1])
    plt.xlabel('$\zeta$')
    plt.ylabel('$\zeta^2 f(\zeta)$')
    #plt.xlim( (0, 5))
    plt.show()

    converteddata=np.transpose(np.array([inputdata[:,0],inputdata[:,1]/inputdata[:,0]**2]))
    convertInterpolated=arrInterpolate(converteddata,1,0)

    plt.plot(converteddata[:,0],converteddata[:,1])
    #plt.yscale("log")

    plt.xlabel('$q$')
    plt.ylabel('$f(q)$')

    #plt.xlim( (0, 5))
    #plt.ylim((1,150))

    plt.show()
    
    plt.plot(convertInterpolated[:,0],convertInterpolated[:,1])
    #plt.yscale("log")

    plt.xlabel('$q$')
    plt.ylabel('$f(q)^{interp}$')
    
    plt.show()

    with open("scalar_{}_converted.dat".format(inputfile), 'w') as f:
                        f.write('#source: https://arxiv.org/pdf/1706.07433.pdf' + '\n' \
                                +'#q=p/T f(q) for r={}'.format(r) + '\n')
                        np.savetxt(f, converteddata, fmt='%4e')

    with open("scalar_{}_converted_interpolated.dat".format(inputfile), 'w') as f:
                        f.write('#source: https://arxiv.org/pdf/1706.07433.pdf' + '\n' \
                                +'#q=p/T f(q) for r={}'.format(r) + '\n')
                        np.savetxt(f, convertInterpolated, fmt='%4e')
```

```python
#Scalars eq spectrum conversion to f(q)

spectrumConversionScalar("scalar_singlet/mydata/distr_eq_mS62.5_x50",50)
```

```python
#Scalars eq spectrum conversion to f(q)

spectrumConversionScalar("scalar_singlet/mydata/distr_noneq_mS62.5_x50",50)
```
